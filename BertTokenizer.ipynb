```python
from transformers import BertTokenizer
import re
```

```python
with open("/content/the_verdict.txt",'r', encoding = 'utf-8') as f:
  raw_text = f.read()
```

```python
raw_text_1 = re.sub(r'\n',' ',raw_text)
raw_text_2 = re.sub(r'--',' ',raw_text_1)
raw_text_3 = re.sub(r'\s+', ' ',raw_text_2)
raw_text_4 = re.sub(r"\\",'',raw_text_3)
text = raw_text_4.strip()
```

```python
# loarding bertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
tokens = tokenizer.encode(text, add_special_tokens=False)
```
