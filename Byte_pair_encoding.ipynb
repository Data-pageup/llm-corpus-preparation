# before encoding we have to clean the dataset

```python
import re
raw_text_1 = re.sub(r'\n',' ',raw_text)
raw_text_2 = re.sub(r'--',' ',raw_text_1)
raw_text_3 = re.sub(r'\s+', ' ',raw_text_2)
raw_text_4 = re.sub(r"\\",'',raw_text_3)
text = raw_text_4.strip()
```
